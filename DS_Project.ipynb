```python
#Load Libraries
import pandas as pd
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot') 

#Load the data
employee_response = pd.read_csv ("ProjectData.csv")
```

```python
#View Data
employee_response.head(5)
```

```python
#Stats
employee_response.describe()
#employee_response.columns
```

```python
#Split the data in meaningful questions and plot

#Question 1 - Order the following factors from most to least relevant
question1 = ['Staff Meals', 'The Mexico/Latino culture', 'Feeling that your contribution makes a difference', 'Sense that everything is in control','When achievements are recognised and celebrated','Music', 'Love for the GYG brand' ]
employee_response[question1].mean().plot(kind='bar')
```

```python
#Question 2 - Select the 5 most important Features
question2 = ['Is an expert in GYG operations','Is encouraging', 'Provides constant feedback', 'Manages themselves well during stressful situations','Truly cares about you and the team', 'Recognises hard work','Develops the team', 'Takes your feedback into consideration', 'Is fair', 'Sets clear expectations','Leads the business and the team to achieving successful results','Works alongside the team as an equal', 'Is polite and courteous','Gives you and the team autonomy to perform your jobs']
employee_response[question2].count().plot(kind='bar')
```

```python
#Question 3 - Select the three most important features
question3 = ['Managers communicate in a disrespectful way','Managers cannot handle stressful situations','Team members are not experienced enough','Team members do not get along','Team members do not help each other', 'The shift is unorganised']
employee_response[question3].count().plot(kind='bar')
```

```python
# fit a linear regression model to verify if features can predict other responses
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
new = ['Manages themselves well during stressful situations', 'Managers cannot handle stressful situations']
temp1 = employee_response [new]
temp=temp1.fillna(0)
feature_cols = ['Manages themselves well during stressful situations']
X = temp[feature_cols]
y = temp['Managers cannot handle stressful situations']
linreg.fit(X, y)
print linreg.intercept_
print linreg.coef_
```

```python
f2_pred = linreg.predict(X)
plt.plot(temp[feature_cols], f2_pred, color='red')
```

#### Data Description

There are 4 questions in the survey used as input data.

Q1: Range in order of importance different factors that affect how happy you feel at work
Q2: Select the 5 most important features that make a good manager
Q3: Select the 3 factors that have the largest effect on making work not enjoyable

The Bar graphs show how this factors relate to each other within each questions.

The logistic regression can be used to predict how selections on one of the questions can affect selections on the others. The one above shows how the selection of 'Manages themselves well during stressful situations' on Q2 can predict the selection of 'Managers cannot handle stressful situations' on Q3.
